### 코딩 목적
한국, 중국, 일본, 베트남, 미국의 실업률 추이를 확인하겠다.

### 필요 데이터

### 코딩 방법
1. 국가별 실업률 데이터 가지고오기(read_csv)
2. 필요 국가 데이터만 분리(데이터 전처리)
3. 그래프 구현(matplotlib)


```python
# 필요 라이브러리 가져오기
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 국가별 실업률 데이터 가져오기
data = pd.read_csv("unemployment analysis.csv")
```

```python
# 데이터 확인 
print("데이터 확인")
print(data)

# Country Name / Country Code / 1991년 ~ 2016년까지 실업률 존재
# 총 235 rows * 33 columns
```

```python
# 필요 국가 데이터만 전처리
data = data[(data["Country Name"] == "China") | (data["Country Name"] == "Korea, Rep.") | (data["Country Name"] == "Japan") | (data["Country Name"] == "Vietnam") | (data["Country Name"] == "United States")]

# 필요없는 Country Code 삭제
del data["Country Code"]

# 데이터 확인 
print("데이터 확인")
print(data)

# 결측치 있는지 확인
print("결측치 확인")
print(data.isnull().sum())

# 확인 결과: 결측치 0
```

```python
# 그래프 그리기위해 데이터만 추출해서 리스트 형식으로 만들기
years = list(data.columns[1:])
chdata = data[data["Country Name"] == "China"].values[0][1:]
kordata = data[data["Country Name"] == "Korea, Rep."].values[0][1:]
jpdata = data[data["Country Name"] == "Japan"].values[0][1:]
vmdata = data[data["Country Name"] == "Vietnam"].values[0][1:]
usdata = data[data["Country Name"] == "United States"].values[0][1:]

print("년도")
print(years)
print("중국 실업률")
print(chdata)
print("한국 실업률")
print(kordata)
print("일본 실업률")
print(jpdata)
print("베트남 실업률")
print(vmdata)
print("미국 실업률")
print(usdata)
```

```python
# 그래프 그리기

# 그래프 사이즈 및 x,y 축 설정
plt.figure(figsize=(10,8))
plt.title('Unemployment Rate by Country')
plt.xlabel("Years")
plt.xticks(rotation=90)
plt.ylabel("Unemployment Rate(%)")

# 데이터 넣기
plt.plot(years, chdata, color="green")
plt.plot(years, kordata, color="black")
plt.plot(years, jpdata, color="red")
plt.plot(years, vmdata, color="yellow")
plt.plot(years, usdata, color="blue")
plt.legend(["China", "Korea", "Japan", "Vietnam", "US"])

# 그래프 보여주기
plt.show()
```
